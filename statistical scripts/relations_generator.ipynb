Load in the powerset function to create a list of all subsets from an iterable

```python
from itertools import chain, combinations

def powerset(iterable):
    """Produces all subsets, including the full set of a given iterable"""
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))
```

Load in the necessary function to query the dataframes for subset relationships

```python
def generate_queries(relations):
    """Generate the pandas dataframe queries for bandgap slicing based on each subset in the relationships input
        Example:
            relations = ["Ketone", "Ether"]
            1. ("Ketone", "NOT Ether")     
            2. ("NOT Ketone", "Ether")     
            3. ("Ketone", "Ether")         
            
        Each output query above can be described as:
            1. Select all structures which have at least 1 ketone and exactly 0 ethers 
            2. Select all structures which have at least 1 ether and exactly 0 ketones
            3. Select all structures which have at least 1 ketone and have at least 1 ether together
    """
    queries = []
    for x in powerset(relations):
        string = ""
        if not x: continue
        for y in relations:
            if y in x:
                temp = " " + y + " > 0 "
                string+=temp
                string+=" & "
            else:
                temp = " " + y + " == 0 "
                string+=temp
                string+=" & "
        string = string[:len(string)-2]
        queries.append(string)
    return queries
```

Create the list of queries relevant for the current subset of functional groups selected

```python
queries = generate_queries(relationships)
print(queries)
```

Computation of bandgap to functional groups bin counts

```python
num_structures = []
for bg in bandgap_values: # loop over bandgap values
    # Slice a bandgap range out of the functional group and bandgap dataframe
    bg_range = bandgap_fgs.loc[
        (bandgap_fgs['bandgap'] >= bg) & 
        (bandgap_fgs['bandgap'] <= bg+0.5)
    ]
    num_structures.append(len(bg_range))    # Checking for validity of 831 total structures over the entire query
    # Insert the total counts of each functional group within that range as the values for a row denoted by bandgap range 
    bin_counts.loc[bg] = [len(bg_range.query(q)) for q in queries] 
```

Computation of the weighted mean of a functional group category. <br> 
Weights $w_i$ are the bandgap values <br> 
$X_i$ are the counts for a specific functional group <br>
$n$ is the number of bins <br>



mean = $\sum_{i=1}^{n}w_i*X_i$/$\sum_{i=1}^{n}w_i$



```python
means = {f:
    round(sum([w*x for (w,x) in zip(bandgap_values, bin_counts[f])])/sum(bin_counts[f]),4) # Mean for each functional group, rounded to 4 decimals
    if (len([x for x in bin_counts[f] if x > 0]) - 1) > 0
    else 0
    for f in cols # loop over every functional group
}
print(means)
```

Computation of standard deviation value of bandgap for a given funcitonal group <br>
$n_w$ is number of bins (AKA number of weights) <br>
$X_i$ is each bin count total <br>
$X_u$ is each functional group bandgap mean <br>
$w_i$ is each bandgap value from the bins

var = $\sum_{i=1}^{n_w}w_i*(X_i-X_u)^2$/$((n_w-1)*\sum_{i=1}^{n}w_i*1/n_w)$ <br> 
<br>
std = $\sqrt{var}$

```python
std = {f:
    round( # Numerator of variance 
        (sum([w*(x - means[f])**2 for (w,x) in zip(bin_counts[f], bandgap_values)])
        / # Denominator of variance
        (((len([x for x in bin_counts[f] if x > 0]) - 1)*sum(bin_counts[f]))/len([x for x in bin_counts[f] if x > 0]))
        )**0.5 # Square root operation on variance
        ,4)
        if (len([x for x in bin_counts[f] if x > 0]) - 1) > 0 # divide by 0 handling
        else
        0
    for f in cols # Loop over all fgs
}
std
```

Append the statistical results to the reusltant matrix

```python
total = bin_counts.copy(deep=True)
total.loc["total"] = list(bin_counts.sum()) 
total.loc["means"] = list(means.values())
total.loc["std"] = list(std.values())
```
